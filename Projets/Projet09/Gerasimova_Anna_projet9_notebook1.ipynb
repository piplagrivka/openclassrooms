```python
#Transformation du table Disponibilité alimentaire
dispo_alimentaire_drop=dispo_alimentaire.drop(columns=['Code Domaine','Domaine','Code zone','Code Élément', 'Code Produit','Code année','Année','Unité','Symbole','Description du Symbole'])
dispo_alimentaire_volailles= dispo_alimentaire_drop.loc[dispo_alimentaire_drop['Produit']=='Viande de Volailles']
#Viande de Volailles
dispo_alimentaire_pivot=dispo_alimentaire_volailles.pivot_table(index='Zone', columns= 'Élément', values='Valeur').reset_index()
dispo_alimentaire_volaille_finale=dispo_alimentaire_pivot.drop(columns=['Alimentation pour touristes',	'Aliments pour animaux','Autres utilisations (non alimentaire)','Pertes','Semences','Traitement','Variation de stock','Résidus'])
dispo_alimentaire_volaille_finale=dispo_alimentaire_volaille_finale.rename(columns ={'Disponibilité alimentaire (Kcal/personne/jour)':'Disponibilité alimentaire de la viande de volaille (Kcal/personne/jour)','Disponibilité alimentaire en quantité (kg/personne/an)':'Disponibilité alimentaire de la viande de volaille en quantité  (kg/personne/an)','Disponibilité de matière grasse en quantité (g/personne/jour)':'Disponibilité de matière grasse de la viande de volaille en quantité (g/personne/jour)','Disponibilité de protéines en quantité (g/personne/jour)':'Disponibilité de protéines de la viande de volaille en quantité (g/personne/jour)','Exportations - Quantité':'Exportations - Quantité (milliers tonnes)','Importations - Quantité':'Importations - Quantité (milliers tonnes)', 'Nourriture':'Nourriture (milliers de tonnes)', 'Production':'Production (milliers de tonnes)', 'Disponibilité intérieure':'Disponibilité intérieure (milliers de tonnes)'})
display(dispo_alimentaire_volaille_finale)
```

```python
#Transformation du table Population
population_2017= population.loc[population['Année']==2017].reset_index()
population_drop= population_2017.drop(columns=['index','Code Domaine','Domaine',	'Code zone', 'Code Élément','Élément','Code Produit','Produit','Code année','Année','Unité','Symbole',	'Description du Symbole',	'Note'])
population_finale= population_drop.rename(columns ={'Valeur':'Population, milles personnes'})
display(population_finale)
#population_finale.to_csv('/content/gdrive/MyDrive/Colab Notebooks/OC_projet09/out.csv')
```

```python
securite_2017= securite.loc[(securite['Année']=='2016-2018') | (securite['Année']=='2017')].reset_index()
securite_drop= securite_2017.drop(columns=['index','Code Domaine','Domaine','Code zone (M49)','Code Élément','Élément','Code Produit','Code année','Année','Unité','Symbole','Description du Symbole','Note'])
securite_pivot=securite_drop.pivot_table(index='Zone', columns= 'Produit', values='Valeur').reset_index()
securite_finale=securite_pivot.drop(columns=['Variabilité de la production alimentaire par habitant ($I par personne constant 2014-16)','Variabilité des disponibilités alimentaires par habitant (Kcal/personne/jour)','Part des céréales, des racines et des tubercules dans les apports énergétiques alimentaires (%) (moyenne sur 3 ans)'])
securite_finale=securite_finale.replace('Pays-Bas (Royaume des)','Pays-Bas')
display(securite_finale.sort_values(by='Zone'))
```

```python
cost_of_food_2017= cost_of_food.loc[cost_of_food['Année']==2017].reset_index()
cost_of_food_drop= cost_of_food_2017.drop(columns=['index','Code Domaine','Domaine','Code zone (M49)','Code Élément','Élément','Code Produit','Code année','Année','Unité','Symbole','Description du Symbole'])
cost_of_food_finale=cost_of_food_drop.pivot_table(index='Zone', columns= 'Produit', values='Valeur').reset_index()
cost_of_food_finale=cost_of_food_finale.replace('Pays-Bas (Royaume des)','Pays-Bas')
cost_of_food_finale= cost_of_food_finale.drop(columns=['Number of people unable to afford a healthy diet (million)'])
display(cost_of_food_finale)
```

```python
macros= pd.read_excel(os.path.join(working_directory, 'macro_finale.xlsx'))
macros_finale=macros.rename(columns ={'Pays':'Zone'})
display(macros_finale)
```

```python
Jointure_1 = pd.merge(population_finale, dispo_alimentaire_volaille_finale, on= 'Zone', how='left')
Jointure_2 = pd.merge(Jointure_1, macros_finale, on= 'Zone', how='left')
Jointure_3 = pd.merge(Jointure_2, cost_of_food_finale, on= 'Zone', how='left')
data_finale = pd.merge(Jointure_3, securite_finale, on= 'Zone', how='left')
display(data_finale)
```

```python
print(data_finale.duplicated().values.any())
print(data_finale.isnull().values.any())
print(data_finale.isnull().sum())
```

```python
data_finale['taux_dependance_import(%)'] = (data_finale['Importations - Quantité (milliers tonnes)'] / data_finale['Disponibilité intérieure (milliers de tonnes)'])*100
data_finale['autosuffisance_prod(%)'] = (data_finale['Production (milliers de tonnes)'] / data_finale['Disponibilité intérieure (milliers de tonnes)'])*100
data_finale_finale = data_finale.drop(columns=['Country Name','Disponibilité intérieure (milliers de tonnes)','Exportations - Quantité (milliers tonnes)','Importations - Quantité (milliers tonnes)','Nourriture (milliers de tonnes)','Production (milliers de tonnes)','Besoins énergétiques alimentaires moyens (kcal/personne/jour)','Valeur des importations alimentaires par rapport aux exportations totales de marchandises (%) (moyenne sur 3 ans)'])
display(data_finale_finale)
```

```python
data_finale_finale = data_finale_finale.fillna(data_finale_finale.mean())
data_finale_finale.isna().mean()
```
